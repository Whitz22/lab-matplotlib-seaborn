#### Plot (x, y) and (x, z).

```python
# Method 1
plt.plot(x,y ,'g^',linewidth=2.0,markersize=4)
plt.title ("X versus Y",fontsize=15)
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

```python
plt.plot(x,z ,'r+',linewidth=2.0,markersize=4)
plt.title ("X versus Z",fontsize=15)
plt.ylabel("Z")
plt.xlabel("X")
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
e = np.exp(x)

plt.figure(figsize=(8, 4))

plt.plot(x, z, color='red',label='x^2')
plt.plot(x, e, color='blue',label='e^x')

plt.xlim(90, 100)

plt.title('Normal Scale')  
plt.savefig('outputfile3.png')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.plot(x, z, color='green',label='x^2')
plt.plot(x, e, color='pink',label='x^2')

plt.title('Logarithmic Scale')

plt.yscale('log')  
plt.xscale('log')  
plt.savefig('outputfile4.png')

plt.legend()

plt.show()
```

```python
# your code here
fitbit=pd.read_csv('fitbit2.csv')
fitbit.head()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_steps = fitbit.groupby('Months_encoded').agg({'Steps':'median'})
median_steps
```

```python

median_steps.plot()
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.title('Median Steps by month')
plt.show()
```

```python
median_steps_Week = fitbit.groupby('Work_or_Weekend').agg({'Steps':'median'})
median_steps_Week
```

```python
median_steps_Week.plot(kind='bar', color=['y']) 

plt.xlabel('Day Type')
plt.xticks(rotation ='horizontal') 
plt.ylabel('Median Steps')
plt.title('Median Steps on Workdays vs. Weekends')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

feature_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

# Create a loop to plot each feature pair
for feature_x, feature_y in feature_pairs:
    plt.figure()  # Create a new figure for each plot
    plt.scatter(fitbit[feature_x], fitbit[feature_y],edgecolor='orange')
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.title(f'{feature_x} vs {feature_y}')


plt.show()
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here

gender = titanic['Gender'].value_counts()

plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="y")
ax.set_title("Bar Graph of Gender", y = 1)
ax.set_xlabel('Gender')
ax.set_ylabel('Number of People')
ax.set_xticklabels(('Male', 'Female'))

```

```python
# Method 2 - seaborn
# your code here
gender=titanic['Gender'].value_counts()
sns.barplot(x = gender.index, y = gender.values,color='pink')

plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.show()
```
